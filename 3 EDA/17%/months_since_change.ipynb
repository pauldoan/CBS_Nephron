```python
import pandas as pd
```

```python
df = pd.read_csv('17.7%.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df['months_since_change'] = df['Date'].where(df['Changed'].eq(1))
```

```python
df['months_since_change'] = (df['Date'].sub(df.groupby('NDC')['months_since_change']
                                        .apply(lambda x: x.shift().ffill())
                                        )
                                        .dt.days
                                        .fillna(0)
                                        .astype(int))/30
```
