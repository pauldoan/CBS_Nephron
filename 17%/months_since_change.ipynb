```python
import pandas as pd
```

```python
df = pd.read_csv('17%.csv')
```

```python
df['Dates'] = pd.to_datetime(df['Dates'], dayfirst=True)
```

```python
df['months_since_change'] = df['Dates'].where(df['Changed'].eq(1))
```

```python
df['months_since_change'] = (df['Dates'].sub(df.groupby('NDC')['months_since_change']
                                        .apply(lambda x: x.shift().ffill())
                                        )
                                        .dt.days
                                        .fillna(0)
                                        .astype(int))/30
```
